```python
import matplotlib.pyplot as plt
from math import comb
```

```python
r=2
p=1/6
```

```python
def neg_binomial_pmf(x,r,p):
  q=1-p
  if x<r:
    return 0
  return comb(x-1,r-1)*(p**r)*(q**(x-r))
x=list(range(1,30))
probs=[neg_binomial_pmf(k,r,p) for k in x]
plt.bar(x,probs)
plt.show()
```

```python
def binomial_distribution(x,r,p):
  q=1-p
  return comb(x,r)*(p**r)*(q**(x-r))

x=list(range(1,30))
probs1=[binomial_distribution(k,r,p) for k in x]

plt.bar(x,probs1)
plt.show()
```
