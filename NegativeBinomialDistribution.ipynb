```python
import matplotlib.pyplot as plt
from math import comb

#parameters
r=2
p=1/6

#PMF
#no on trails required for fixed no.of success
def neg_binomial_pmf(x,r,p): #x is no of event, r is no of success
  if x<r:
    return 0
  q=1-p
  return comb(x-1,r-1)*(p**r)*(q**(x-r))

#values
x=list(range(1,30))
probs=[neg_binomial_pmf(k,r,p) for k in x]

#plot
plt.bar(x,probs)
plt.show()
```
